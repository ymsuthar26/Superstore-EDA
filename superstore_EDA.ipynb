```python
city_profit=df.groupby('City')['Profit'].mean()
top_profit=city_profit.sort_values(ascending=False).head(10)
top_loss=city_profit.sort_values().head(10)

plt.bar(top_profit.index,top_profit.values)
plt.title('top 10 city making profit')
plt.xlabel('city')
plt.ylabel("avg profit")
plt.xticks(rotation=45)
plt.show()

plt.bar(top_loss.index,top_loss.values)
plt.title('top 10 city making loss')
plt.xlabel('city')
plt.ylabel("avg loss")
plt.xticks(rotation=45)
plt.show()
```

```python
top_sales=df.groupby('City')['Sales'].sum().sort_values(ascending=False).head(10)
plt.bar(top_sales.index,top_sales.values)
plt.title('city wise total sales')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(50,20))
sns.boxplot(data=df,x="City",y="Sales")
plt.title("sales distrubution across cities")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
order_value=df.groupby(['City','Order ID'])['Sales'].sum().groupby('City').mean().sort_values(ascending=False).head(10)
```
